### Напишите скрипт, который проходит циклом от 1 до 200. Если цифра чётная, то её необходимо добавить в список - list1, если нечётная - list2, если делиться на 3 без остатка - list3, если делиться на 5 без остатка - list4. Найдите пересечения значений в list1, list3 и list4, если таковые имеются, и выведите их

```python
# code
list1 = []
list2 = []
list3 = []
list4 = []

for i in range(1, 201):
    if i % 2 == 0:
        list1.append(i)
    if i % 2 != 0:
        list2.append(i)
    if i % 3 == 0:
        list3.append(i)
    if i % 5 == 0:
        list4.append(i)

intersec = set(list1) & set(list3) & set(list4)
print(intersec)
```

```python
# code
class Animal:
    count = 0

    def __init__(self, name, legs, sound):
        self.name = name
        self.legs = legs
        self.sound = sound
        Animal.count += 1

    def make_sound(self):
        for _ in range(4):
            print(self.sound)

    @staticmethod
    def get_total_count():
        return Animal.count
```
